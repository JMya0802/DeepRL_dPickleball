```python
yellow = np.array([255, 255, 45])
diff_map = np.zeros(img.shape)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        diff_map[i, j] = np.linalg.norm(yellow - img[i, j]) # L2 distance from yellow

diff_map = diff_map / np.max(diff_map) # normalization
binary_img = diff_map < 0.15 # thresholding
# print(diff_map[40:60, 80:100][9])
plt.imshow(binary_img.astype(np.uint8) * 255)
# plt.imshow(diff_map)
```
